```python
import cv2
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread('/content/image1.jpg',0)
```

```python
#calculate histogram
histr = cv2.calcHist([img],[0],None,[256],[0,256])
```

```python
plt.plot(histr)
```

```python
plt.show()
```

```python
#transpose image
window_name = 'IMAGE'
image = cv2.transpose(img)
cv2_imshow(image)
```
